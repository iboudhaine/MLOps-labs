##### `javap` Analysis: First `Order` Class

*   **Private, Immutable Fields:** The constructor parameters (`orderId`, `orderDate`, etc.) are compiled into `private final` fields.
    *   **`private`**: They are completely hidden and cannot be accessed from outside the class.
    *   **`final`**: They are immutable and cannot be changed after the object is created.

*   **No Getters:** The compiler does not generate any public methods to read these private fields.


Let's create an Order object and try to access one of its attribute:

````python
%%bash
cat code/exercice2.scala || true
````

````python
%%bash
scalac code/exercice2.scala || true
````

We got the expected error: `value orderId is not a member of Order`.

**Conclusion:** Parameters without `val` or `var` are treated as temporary, internal values needed only to construct the object. They are not public properties. This is why `order.orderId` fails.


Let us now add `val` before all constructor arguments.

````python
%%bash
cat code/exercice2.scala || true
````

````python
%%bash
scalac code/exercice2.scala || true
javap -p Order || true
````

We notice that the fields became permanent, public, read-only part of the object's structure.

**Conclusion:** By adding `val`, we instruct the Scala compiler to automatically create a public, immutable field for that parameter.


Let's do the same test as before (access an attribute), and see what will happen.

````python
%%bash
cat code/exercice2.scala || true
````

````python
%%bash
scalac code/exercice2.scala || true
scala Main || true
````

See, now it compiles without errors.

Let's try modifying the attribute now...

````python
%%bash
cat code/exercice2.scala || true
````

````python
%%bash
scalac code/exercice2.scala || true
````

The process fails. Apparently the compiler enforces the immutability of a `val`. The attempt to reassign it does not seem to work.

Let's try `var` instead...

````python
%%bash
cat code/exercice2.scala || true
````

````python
%%bash
scalac code/exercice2.scala || true
scala Main || true
````

Looks like it worked. By changing `val` to `var` in the constructor, we are instructing the Scala compiler to generate both an accessor (a getter) and a mutator (a setter) for that field.

**Conclusion**:
- `val` defines an immutable value, a constant that cannot be reassigned.
- `var` defines a mutable variable, for which the compiler automatically generates the necessary setter method to allow modification.

##### Step 1: Let's Create the SBT Project Structure
Let's:
1. Create the main project folder and enter it
2. Create the nested source directory

````python
%%bash
mkdir code/OrderRevenuePrice
cd code/OrderRevenuePrice
mkdir -p src/main/scala
````

##### Step 2: Create the SBT Build File

SBT needs a configuration file in the project's root directory to know the project's name, version, and Scala version.

We'll create a file named `build.sbt` in the `OrderRevenuePrice` root folder with the following content:

**File: `build.sbt`**
```sbt
name := "OrderRevenuePrice"
version := "0.1"
scalaVersion := "2.11.0"
```


````python
%%bash
cat code/OrderRevenuePrice/build.sbt
````

##### Step 3:

1.  **Move the Data:** We'll move `part-00000` file into the `OrderRevenuePrice` root folder.
2.  **Create the Scala File:** Inside the `src/main/scala/` directory, we'll create the file `OrderRevenue.scala` with the initial code from the exercise 3.


````python
%%bash
cat code/OrderRevenuePrice/src/main/scala/OrderRevenue.scala
````

````python
%%bash
cd code/OrderRevenuePrice
sbt "runMain OrderRevenue"
````
